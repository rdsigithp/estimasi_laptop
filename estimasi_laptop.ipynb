# seleksi fitur

```python
features = ['graphic_card_gb','warranty','old_price','discount','star_rating']
x = df[features]
y = df['latest_price']
x.shape, y.shape
```

# split data training dan data testing

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

# membuat model regresi linier

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(X_test)
```

```python
score = lr.score(X_test, y_test)
print('akurasi model regresi linier = ', score)
```

# membuat inputan model regresi linier

```python
#['graphic_card_gb','warranty','old_price','discount','star_rating']
input_data = np.array([[2,1,111019,26,4.4]])

prediction = lr.predict(input_data)
print('Estimasi harga Laptop dalam RUPEE :', prediction)
```
